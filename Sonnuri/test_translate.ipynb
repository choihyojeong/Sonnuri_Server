```python
import websocket
import cv2
import json
import time
import threading

def send_frames(ws):
    cap = cv2.VideoCapture(0)  # 웹캠 열기
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        _, buffer = cv2.imencode('.jpg', frame)
        ws.send(buffer.tobytes(), opcode=websocket.ABNF.OPCODE_BINARY)
        time.sleep(0.1)  # 10fps로 전송 속도 조절
    cap.release()
    ws.close()

def on_message(ws, message):
    try:
        result = json.loads(message)
        char = result.get("char")
        confidence = result.get("confidence")
        accepted = result.get("accepted")
        current_word = result.get("current_word")
        print(f"[결과] 자모: {char}, 정확도: {confidence}, 채택됨: {accepted}, 현재 단어: {current_word}")
    except Exception as e:
        print("[메시지 파싱 오류]", e)

def on_error(ws, error):
    print("[에러 발생]", error)

def on_close(ws, close_status_code, close_msg):
    print("[연결 종료]")

def on_open(ws):
    print("[연결 시작]")
    threading.Thread(target=send_frames, args=(ws,)).start()

# 번역 웹소켓 URL
ws_url = "ws://localhost:8000/stream/translate"

ws = websocket.WebSocketApp(
    ws_url,
    on_open=on_open,
    on_message=on_message,
    on_error=on_error,
    on_close=on_close
)

ws.run_forever()
```
